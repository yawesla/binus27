##### **Bisection**

###### Kalo ada 2 titik A sama B dimana salah satunya positif dan satunya negatif akan ada titik c yang nilainya 0 diantaranya

```python
import numpy as np
# X^2-2x-3
def f(x):
    return x**2 - 2*x -3

def bisection(a,b, tolerance):
    # hasil dari f(x)nya, yg satu harus positif yg satu ngetif
    # 1 -> positif
    # 0 -> negatif
    if np.sign(f(a)) == np.sign(f(b)):
        print("Coordinate Invalid")
        return 
    
    c = (a + b) / 2

    if(np.abs(f(c)) < tolerance):
        print("Root = ",c)
        return 
    elif(np.sign(f(c)) == np.sign(f(b))):
        bisection(a,c,tolerance)
    elif(np.sign(f(c)) == np.sign(f(a))):
        bisection(c,b,tolerance) 

bisection(0,5,0.015)
bisection(0,1,0.015)
```

##### **Newton-Raphson**

###### Rumus : x1 = x0 - f(x0)/f`(x0)

```python
import numpy as np

def f(x):
    return x**2 - 2*x -3

def g(x):
    return 2*x -2

def newton_raphson(x0,tolerance):
    x1 = x0 - f(x0) / g(x0)

    if np.abs(f(x1)) < tolerance:
        print("Root = ",x1)
        return
    
    newton_raphson(x1,tolerance)

newton_raphson(2,0.015)
```
